```python
import xarray as xr
import numpy as np

# Function to compute gradient magnitude
def compute_gradient_magnitude(data):
    # Calculate gradients along i and j dimensions
    di = np.gradient(data, axis=1)  # i dimension
    dj = np.gradient(data, axis=2)  # j dimension
    # Calculate magnitude of gradient
    grad_mag = np.sqrt(di**2 + dj**2)
    return grad_mag

# Function to normalize SSH by its range
def normalize_ssh(ssh_data):
    # Compute the max range for each sample
    max_range = ssh_data.max(dim=('i', 'j')) - ssh_data.min(dim=('i', 'j'))
    # Normalize by dividing by half the range (to get range between -1 and 1)
    normalized = ssh_data / (max_range).values[:, np.newaxis, np.newaxis]
    return normalized

# Vectorized function to normalize UBM by the std of SSH gradient magnitude
def normalize_ubm_vectorized(ubm_data, ssh_data):
    # Replace NaN values if any
    ubm_data = ubm_data.fillna(0)
    
    # Extract all SSH fields at once
    ssh_values = ssh_data.values
    
    # Compute gradient magnitude for all samples at once
    grad_mag = compute_gradient_magnitude(ssh_values)
    
    # Compute std of gradient magnitude for each sample
    std_grad_mag = np.std(grad_mag, axis=(1, 2))
    
    # Add small epsilon to avoid division by zero
    epsilon = 1e-8
    std_grad_mag = std_grad_mag + epsilon
    
    # Reshape std_grad_mag for broadcasting
    std_grad_mag_reshaped = std_grad_mag[:, np.newaxis, np.newaxis]
    
    # Create new DataArray with the normalized values
    normalized_ubm = ubm_data / std_grad_mag_reshaped
    
    return normalized_ubm

# Process all datasets
def normalize_datasets(train_ds, val_ds, test_ds):
    # Function to normalize a single dataset
    def normalize_data_fast(ds):
        # Create a new dataset for normalized data
        ds_normalized = ds.copy()
        
        # Replace NaN values in UBM data
        if 'ubm_A' in ds:
            ds['ubm_A'] = ds.ubm_A.fillna(0)
        if 'ubm_B' in ds:
            ds['ubm_B'] = ds.ubm_B.fillna(0)
        
        # Normalize Region A
        ds_normalized['ssh_A_normalized'] = normalize_ssh(ds.ssh_A)
        ds_normalized['ubm_A_normalized'] = normalize_ubm_vectorized(ds.ubm_A, ds.ssh_A)
        
        # Normalize Region B
        ds_normalized['ssh_B_normalized'] = normalize_ssh(ds.ssh_B)
        ds_normalized['ubm_B_normalized'] = normalize_ubm_vectorized(ds.ubm_B, ds.ssh_B)
        
        return ds_normalized
    
    # Apply normalization to all datasets
    print("Normalizing training data...")
    train_normalized = normalize_data_fast(train_ds)
    
    print("Normalizing validation data...")
    val_normalized = normalize_data_fast(val_ds)
    
    print("Normalizing test data...")
    test_normalized = normalize_data_fast(test_ds)
    
    return train_normalized, val_normalized, test_normalized

# Apply normalization to all datasets
normalized_train, normalized_val, normalized_test = normalize_datasets(train, val, test)
```

# Training

```python
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.cuda.manual_seed_all(42)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    

loss_train = MseLoss()
loss_val = MseLoss()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = UNet(in_channels=1, out_channels=1, initial_features=32, depth=4)
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

def train_model(model, train_loader, val_loader,
                loss_train, loss_val,
                optimizer, device,
                probabilistic=False, 
                save_path='/home/jovyan/SSH/B_data/updated_dm/test3/model.pth',
                n_epochs=2000,
                patience=50):

    """
    This function trains a deep learning model.
    
    Parameters:
    - model: The neural network model to be trained
    - train_loader: DataLoader for training data
    - val_loader: DataLoader for validation data
    - loss_train: Loss function for training
    - loss_val: Loss function for validation
    - optimizer: Optimization algorithm
    - device: Device to run the model on (CPU or GPU)
    - probabilistic: Whether the model is probabilistic
    - save_path: Path to save the trained model
    - n_epochs: Number of training epochs
    - patience: Number of epochs to wait for improvement in validation loss
                before early stopping.
    """

    model.to(device)
    
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)
    
    # Variables for early stopping and loss tracking
    best_val_loss = float('inf')
    patience_counter = 0
    train_losses = []
    val_losses = []
    
    # Check if a saved model exists and load it if it does
    if os.path.isfile(save_path):
        checkpoint = torch.load(save_path)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
        start_epoch = checkpoint['epoch'] + 1
        best_val_loss = checkpoint['best_val_loss']
        patience_counter = checkpoint['patience_counter']
        train_losses = checkpoint.get('train_losses', [])
        val_losses = checkpoint.get('val_losses', [])
        print(f"Resuming from epoch {start_epoch} with best_val_loss = {best_val_loss:.3e}")
    else:
        start_epoch = 0

    # Main training loop
    for epoch in range(start_epoch, n_epochs):
        
        start_time = time.time()
        model.train()  # Set model to training mode
        train_running_loss = 0.0

        # Training phase
        for batch_x, batch_y in train_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)
            
            # Forward pass
            outputs = model(batch_x)
            loss = loss_train(outputs, batch_y)
            
            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            train_running_loss += loss.item() * batch_x.size(0)
            
        # Calculate average training loss for the epoch
        epoch_loss = train_running_loss / len(train_loader.dataset)
        train_losses.append(epoch_loss)

        # Validation phase
        model.eval()
        val_running_loss = 0.0
        with torch.no_grad():
            for batch_x, batch_y in val_loader:
                batch_x = batch_x.to(device)
                batch_y = batch_y.to(device)
                outputs = model(batch_x)
                
                # Handle probabilistic case
                if probabilistic:
                    # For probabilistic case, select the mean (first channel)
                    outputs = outputs[:, 0, ...].unsqueeze(1)
                
                # Compute validation loss directly without ZCA transform
                loss = loss_val(outputs, batch_y)
                val_running_loss += loss.item() * batch_x.size(0)
                
        val_loss = val_running_loss / len(val_loader.dataset)
        val_losses.append(val_loss)
        
        # Update learning rate based on validation loss
        scheduler.step(val_loss)

        # Calculate epoch duration
        end_time = time.time()
        epoch_duration = end_time - start_time

        print(f"Epoch {epoch+1}, Train Loss: {epoch_loss:.2e}, Val Loss: {val_loss:.2e}, "
              f"Epoch Time: {epoch_duration:.2f}s")

        # Early stopping
        if val_loss < 1.02*best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0
            # Save checkpoint only if we have a new best validation loss
            checkpoint = {
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'scheduler_state_dict': scheduler.state_dict(),
                'best_val_loss': best_val_loss,
                'patience_counter': patience_counter,
                'train_losses': train_losses,
                'val_losses': val_losses
            }
            torch.save(checkpoint, save_path)
            print(f"Best model so far saved at epoch {epoch+1} (Val Loss: {best_val_loss:.3e})")
        else:
            patience_counter += 1
            print(f"Patience counter: {patience_counter}/{patience}")
        
        # Early stopping check
        if patience_counter >= patience:
            print("Early stopping triggered.")
            break

    print("Training complete")



# train_model(model, train_loader, val_loader,
#             loss_train, loss_val,
#             optimizer, device,
#             probabilistic=False,
#             save_path ='/home/jovyan/GRL/checkpoint/regionA_norm.pth',
#             n_epochs = 200, 
#             patience = 50)
```

```python
import os
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader

def evaluate_models(model_class, model_params, device, normalized_test, test, checkpoint_paths, probabilistic=False):
    """
    Evaluate both normalized and unnormalized models together
    
    Parameters:
    - model_class: The neural network model class to instantiate
    - model_params: Parameters to pass to the model constructor
    - device: Device to run the model on (CPU or GPU)
    - normalized_test: Dataset with normalized values
    - test: Dataset with unnormalized values
    - checkpoint_paths: Dictionary with 'norm' and 'nonorm' paths to the saved model checkpoints
    - probabilistic: Whether the models are probabilistic
    
    Returns:
    - Dictionary with evaluation results for both models
    """
    results = {}
    batch_size = 64
    eps = 1e-6
    
    # 1. First evaluate the normalized model
    print("Evaluating normalized model...")
    # Prepare normalized data loader
    x_test_norm = torch.from_numpy(normalized_test.ssh_B_normalized.values).float().unsqueeze(1).to(device)
    y_test_norm = torch.from_numpy(normalized_test.ubm_B_normalized.values).float().unsqueeze(1).to(device)
    norm_test_dataset = TensorDataset(x_test_norm, y_test_norm)
    norm_test_loader = DataLoader(norm_test_dataset, batch_size=batch_size, shuffle=False)
    
    # Create and load normalized model
    norm_model = model_class(**model_params).to(device)
    norm_model.eval()
    
    if os.path.isfile(checkpoint_paths['norm']):
        checkpoint = torch.load(checkpoint_paths['norm'], map_location=device)
        norm_model.load_state_dict(checkpoint['model_state_dict'])
        print(f"Loaded normalized model from {checkpoint_paths['norm']}")
    else:
        raise FileNotFoundError(f"No checkpoint at {checkpoint_paths['norm']}")
    
    # Collect predictions for normalized model
    norm_prediction_list = []
    norm_sigma_list = []
    
    with torch.no_grad():
        for batch_x, _ in norm_test_loader:
            batch_x = batch_x.to(device)
            outputs = norm_model(batch_x)
            
            if probabilistic:
                mu = outputs[:, 0, ...]
                log_sigma = outputs[:, 1, ...]
                sigma = torch.exp(log_sigma).clamp(min=eps)
                norm_prediction_list.append(mu)
                norm_sigma_list.append(sigma)
            else:
                norm_prediction_list.append(outputs)
    
    # Process normalized model outputs
    norm_predictions = torch.cat(norm_prediction_list, dim=0)
    norm_predictions_np = norm_predictions.cpu().numpy().squeeze(1)
    
    # Denormalize predictions
    denormalized_predictions = denormalize_ubm(norm_predictions_np, test.ssh_B.values)
    
    # Process uncertainty if probabilistic
    if probabilistic:
        norm_sigma = torch.cat(norm_sigma_list, dim=0)
        norm_sigma_np = norm_sigma.cpu().numpy()
        denormalized_sigma = denormalize_ubm(norm_sigma_np, test.ssh_B.values)
        norm_sigma_da = xr_da(denormalized_sigma, test.ssh_B)
        results['norm_sigma'] = norm_sigma_da
    
    # Convert to xarray DataArrays
    norm_prediction_da = xr_da(denormalized_predictions, test.ssh_B)
    norm_bm_da = test.ssh_B - norm_prediction_da
    
    # Store results
    results['norm_ubm'] = norm_prediction_da
    results['norm_bm'] = norm_bm_da
    
    # 2. Now evaluate the unnormalized model
    print("Evaluating unnormalized model...")
    # Prepare unnormalized data loader
    x_test_unnorm = torch.from_numpy(test.ssh_B.values).float().unsqueeze(1).to(device)
    y_test_unnorm = torch.from_numpy(test.ubm_B.values).float().unsqueeze(1).to(device)
    unnorm_test_dataset = TensorDataset(x_test_unnorm, y_test_unnorm)
    unnorm_test_loader = DataLoader(unnorm_test_dataset, batch_size=batch_size, shuffle=False)
    
    # Create and load unnormalized model
    unnorm_model = model_class(**model_params).to(device)
    unnorm_model.eval()
    
    if os.path.isfile(checkpoint_paths['nonorm']):
        checkpoint = torch.load(checkpoint_paths['nonorm'], map_location=device)
        unnorm_model.load_state_dict(checkpoint['model_state_dict'])
        print(f"Loaded unnormalized model from {checkpoint_paths['nonorm']}")
    else:
        raise FileNotFoundError(f"No checkpoint at {checkpoint_paths['nonorm']}")
    
    # Collect predictions for unnormalized model
    unnorm_prediction_list = []
    unnorm_sigma_list = []
    
    with torch.no_grad():
        for batch_x, _ in unnorm_test_loader:
            batch_x = batch_x.to(device)
            outputs = unnorm_model(batch_x)
            
            if probabilistic:
                mu = outputs[:, 0, ...]
                log_sigma = outputs[:, 1, ...]
                sigma = torch.exp(log_sigma).clamp(min=eps)
                unnorm_prediction_list.append(mu)
                unnorm_sigma_list.append(sigma)
            else:
                unnorm_prediction_list.append(outputs)
    
    # Process unnormalized model outputs
    unnorm_predictions = torch.cat(unnorm_prediction_list, dim=0)
    unnorm_predictions_np = unnorm_predictions.cpu().numpy().squeeze(1)
    
    # Process uncertainty if probabilistic
    if probabilistic:
        unnorm_sigma = torch.cat(unnorm_sigma_list, dim=0)
        unnorm_sigma_np = unnorm_sigma.cpu().numpy()
        unnorm_sigma_da = xr_da(unnorm_sigma_np, test.ssh_B)
        results['unnorm_sigma'] = unnorm_sigma_da
    
    # Convert to xarray DataArrays
    unnorm_prediction_da = xr_da(unnorm_predictions_np, test.ssh_B)
    unnorm_bm_da = test.ssh_B - unnorm_prediction_da
    
    # Store results
    results['unnorm_ubm'] = unnorm_prediction_da
    results['unnorm_bm'] = unnorm_bm_da
    
    return results

def denormalize_ubm(normalized_ubm, ssh_data):
    """
    Denormalize UBM predictions by multiplying by the standard deviation of SSH gradient magnitude
    
    Parameters:
    - normalized_ubm: Normalized UBM predictions from the model
    - ssh_data: Original SSH data used for normalization
    
    Returns:
    - denormalized_ubm: UBM values in original scale
    """
    # Compute gradient magnitude
    grad_mag = compute_gradient_magnitude(ssh_data)
    
    # Compute std of gradient magnitude for each sample
    std_grad_mag = np.std(grad_mag, axis=(1, 2))
    
    # Add small epsilon to avoid division by zero (same as in normalization)
    epsilon = 1e-8
    std_grad_mag = std_grad_mag + epsilon
    
    # Reshape std_grad_mag for broadcasting
    std_grad_mag_reshaped = std_grad_mag[:, np.newaxis, np.newaxis]
    
    # Multiply normalized UBM by std_grad_mag to denormalize
    denormalized_ubm = normalized_ubm * std_grad_mag_reshaped
    
    return denormalized_ubm

# Usage example
checkpoint_paths = {
    'norm': '/home/jovyan/GRL/checkpoint/regionA_norm.pth',
    'nonorm': '/home/jovyan/GRL/checkpoint/regionA_nonorm.pth'
}

model_params = {
    'in_channels': 1,
    'out_channels': 1,
    'initial_features': 32,
    'depth': 4
}

# Run evaluation for both models
results = evaluate_models(
    model_class=UNet,
    model_params=model_params,
    device=device,
    normalized_test=normalized_test,
    test=test,
    checkpoint_paths=checkpoint_paths,
    probabilistic=False
)

# Now you can access results for both models
norm_ubm = results['norm_ubm']      # Denormalized predictions from normalized model
norm_bm = results['norm_bm']       
unnorm_ubm = results['unnorm_ubm']  # Predictions from unnormalized model
unnorm_bm = results['unnorm_bm']   

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.ndimage import gaussian_filter1d

def calculate_rmse(prediction, ground_truth):
    """Calculate RMSE for each sample between prediction and ground truth"""
    # Ensure we're working with numpy arrays
    if hasattr(prediction, 'values'):
        prediction = prediction.values
    if hasattr(ground_truth, 'values'):
        ground_truth = ground_truth.values
    
    # Initialize array to store RMSE for each sample
    num_samples = prediction.shape[0]
    rmse_values = np.zeros(num_samples)
    
    # Calculate RMSE for each sample
    for i in range(num_samples):
        # Flatten the 2D spatial dimensions
        pred_flat = prediction[i].flatten()
        truth_flat = ground_truth[i].flatten()
        
        # Remove NaN values if any
        mask = ~np.isnan(pred_flat) & ~np.isnan(truth_flat)
        pred_clean = pred_flat[mask]
        truth_clean = truth_flat[mask]
        
        # Calculate RMSE
        rmse_values[i] = np.sqrt(mean_squared_error(truth_clean, pred_clean))
    
    return rmse_values

def smooth_data(data, sigma=2):
    """Apply Gaussian smoothing to data"""
    return gaussian_filter1d(data, sigma=sigma)

def compare_bm_performance_log_scale_plot(results, test_data, sigma=3):
    """
    Compare the performance of normalized and unnormalized BM models using smooth line plots with log scale
    
    Parameters:
    - results: Dictionary with model predictions
    - test_data: Test dataset with ground truth values
    - sigma: Standard deviation for Gaussian smoothing
    """
    # Calculate RMSE for BM
    ground_truth_bm = test_data.ssh_B - test_data.ubm_B
    norm_bm_rmse = calculate_rmse(results['norm_bm'], ground_truth_bm)
    unnorm_bm_rmse = calculate_rmse(results['unnorm_bm'], ground_truth_bm)
    
    # Apply smoothing
    norm_bm_rmse_smooth = smooth_data(norm_bm_rmse, sigma)
    unnorm_bm_rmse_smooth = smooth_data(unnorm_bm_rmse, sigma)
    
    # Get number of samples
    num_samples = len(norm_bm_rmse)
    sample_indices = np.arange(num_samples)
    
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Plot BM RMSE comparison - only smooth lines
    plt.plot(sample_indices, norm_bm_rmse_smooth, 'g-', linewidth=2.5, label='Normalized')
    plt.plot(sample_indices, unnorm_bm_rmse_smooth, 'm-', linewidth=2.5, label='Unnormalized')
    
    # plt.title('BM RMSE Comparison', fontsize=14)
    plt.xlabel('Testing Samples', fontsize=15)
    plt.ylabel('RMSE (m)', fontsize=15)
    plt.legend(loc='best')
    
    # Add horizontal lines for mean values
    norm_bm_mean = np.mean(norm_bm_rmse)
    unnorm_bm_mean = np.mean(unnorm_bm_rmse)
    
    # Add text annotations for mean values - adjust position for log scale
    y_min, y_max = plt.ylim()
    text_y_pos = np.exp(np.log(y_min) + 0.1 * (np.log(y_max) - np.log(y_min)))
    
    plt.text(num_samples * 0.02, text_y_pos * 2, 
            f'Norm Mean: {norm_bm_mean:.3e}', 
            color='green', fontsize=18)
    plt.text(num_samples * 0.02, text_y_pos * 1.8, 
            f'Unnorm Mean: {unnorm_bm_mean:.3e}', 
            color='magenta', fontsize=18)
    
    # Print statistics
    print("\nBM RMSE Statistics:")
    print(f"  Normalized Model - Mean: {norm_bm_mean:.4f}, Median: {np.median(norm_bm_rmse):.4f}")
    print(f"  Unnormalized Model - Mean: {unnorm_bm_mean:.4f}, Median: {np.median(unnorm_bm_rmse):.4f}")
    print(f"  Improvement: {(unnorm_bm_mean - norm_bm_mean)/unnorm_bm_mean*100:.2f}%")
    
    # Finalize layout
    plt.tight_layout()
    # plt.savefig('bm_performance_log_scale_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'bm_rmse': {
            'norm': norm_bm_rmse,
            'unnorm': unnorm_bm_rmse,
            'norm_smooth': norm_bm_rmse_smooth,
            'unnorm_smooth': unnorm_bm_rmse_smooth
        }
    }

# Usage example
performance_stats = compare_bm_performance_log_scale_plot(
    results, 
    test,
    sigma=15  # Adjust sigma for Gaussian filter strength
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.ndimage import gaussian_filter1d
import xrft
import random
import xarray as xr
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

def calculate_rmse(prediction, ground_truth):
    """Calculate RMSE for each sample between prediction and ground truth"""
    # Ensure we're working with numpy arrays
    if hasattr(prediction, 'values'):
        prediction = prediction.values
    if hasattr(ground_truth, 'values'):
        ground_truth = ground_truth.values
    
    # Initialize array to store RMSE for each sample
    num_samples = prediction.shape[0]
    rmse_values = np.zeros(num_samples)
    
    # Calculate RMSE for each sample
    for i in range(num_samples):
        # Flatten the 2D spatial dimensions
        pred_flat = prediction[i].flatten()
        truth_flat = ground_truth[i].flatten()
        
        # Remove NaN values if any
        mask = ~np.isnan(pred_flat) & ~np.isnan(truth_flat)
        pred_clean = pred_flat[mask]
        truth_clean = truth_flat[mask]
        
        # Calculate RMSE
        rmse_values[i] = np.sqrt(mean_squared_error(truth_clean, pred_clean))
    
    return rmse_values

def smooth_data(data, sigma=2):
    """Apply Gaussian smoothing to data"""
    return gaussian_filter1d(data, sigma=sigma)

def isotropic_spectra(data):
    """
    Calculate the isotropic power spectrum of input data.
    
    Parameters:
    -----------
    data : xarray.DataArray
        Input data array with dimensions 'i' and 'j'.
    
    Returns:
    --------
    xarray.DataArray
        The isotropic power spectrum calculated using xrft.
    """
    # Replace NaN values with zeros
    data_no_nan = data.fillna(0)
    
    iso_psd = xrft.isotropic_power_spectrum(
        data_no_nan, 
        dim=['i', 'j'], 
        detrend='constant', 
        window=True
    )
    return iso_psd

def compare_psd_and_rmse(results, test_data, n_samples=100, sigma=15, random_seed=42):
    """
    Compare the PSD of different models and plot RMSE comparison
    
    Parameters:
    -----------
    results : dict
        Dictionary containing model predictions
    test_data : xarray.Dataset
        Test dataset with ground truth values
    n_samples : int
        Number of random samples to select
    sigma : float
        Standard deviation for Gaussian smoothing
    random_seed : int
        Random seed for reproducibility
    """
    # Set random seed
    np.random.seed(random_seed)
    random.seed(random_seed)
    
    # Get total number of samples
    total_samples = results['norm_ubm'].shape[0]
    
    # Randomly select samples
    random_indices = np.random.choice(total_samples, size=n_samples, replace=False)
    random_indices.sort()  # Sort for consistent ordering
    
    # Select random samples
    norm_ubm_selected = results['norm_ubm'].isel(sample=random_indices)
    unnorm_ubm_selected = results['unnorm_ubm'].isel(sample=random_indices)
    ground_truth_ubm_selected = test_data.ubm_B.isel(sample=random_indices)
    
    # Calculate BM data
    ground_truth_bm = test_data.ssh_B - test_data.ubm_B
    norm_bm_selected = results['norm_bm'].isel(sample=random_indices)
    unnorm_bm_selected = results['unnorm_bm'].isel(sample=random_indices)
    ground_truth_bm_selected = ground_truth_bm.isel(sample=random_indices)
    
    # Replace NaN values with zeros in all datasets
    norm_ubm_selected = norm_ubm_selected.fillna(0)
    unnorm_ubm_selected = unnorm_ubm_selected.fillna(0)
    ground_truth_ubm_selected = ground_truth_ubm_selected.fillna(0)
    norm_bm_selected = norm_bm_selected.fillna(0)
    unnorm_bm_selected = unnorm_bm_selected.fillna(0)
    ground_truth_bm_selected = ground_truth_bm_selected.fillna(0)
    
    # Calculate PSD for each sample and then average
    print("Calculating PSDs for UBM data...")
    norm_ubm_psd_list = []
    unnorm_ubm_psd_list = []
    ground_truth_ubm_psd_list = []
    
    for i in range(n_samples):
        norm_ubm_psd_list.append(isotropic_spectra(norm_ubm_selected.isel(sample=i)))
        unnorm_ubm_psd_list.append(isotropic_spectra(unnorm_ubm_selected.isel(sample=i)))
        ground_truth_ubm_psd_list.append(isotropic_spectra(ground_truth_ubm_selected.isel(sample=i)))
    
    # Average PSDs for UBM
    norm_ubm_psd_avg = xr.concat(norm_ubm_psd_list, dim='sample').mean(dim='sample')
    unnorm_ubm_psd_avg = xr.concat(unnorm_ubm_psd_list, dim='sample').mean(dim='sample')
    ground_truth_ubm_psd_avg = xr.concat(ground_truth_ubm_psd_list, dim='sample').mean(dim='sample')
    
    print("Calculating PSDs for BM data...")
    norm_bm_psd_list = []
    unnorm_bm_psd_list = []
    ground_truth_bm_psd_list = []
    
    for i in range(n_samples):
        norm_bm_psd_list.append(isotropic_spectra(norm_bm_selected.isel(sample=i)))
        unnorm_bm_psd_list.append(isotropic_spectra(unnorm_bm_selected.isel(sample=i)))
        ground_truth_bm_psd_list.append(isotropic_spectra(ground_truth_bm_selected.isel(sample=i)))
    
    # Average PSDs for BM
    norm_bm_psd_avg = xr.concat(norm_bm_psd_list, dim='sample').mean(dim='sample')
    unnorm_bm_psd_avg = xr.concat(unnorm_bm_psd_list, dim='sample').mean(dim='sample')
    ground_truth_bm_psd_avg = xr.concat(ground_truth_bm_psd_list, dim='sample').mean(dim='sample')
    
    # Create figure with three subplots
    fig, axes = plt.subplots(1, 3, figsize=(24, 7))
    
    # Find global y-axis limits
    all_psd_data = [
        norm_ubm_psd_avg, unnorm_ubm_psd_avg, ground_truth_ubm_psd_avg,
        norm_bm_psd_avg, unnorm_bm_psd_avg, ground_truth_bm_psd_avg
    ]
    
    global_ymin = min([psd.min().values for psd in all_psd_data if not np.isnan(psd.min().values)])
    global_ymax = max([psd.max().values for psd in all_psd_data if not np.isnan(psd.max().values)])
    
    # Ensure positive values for log scale
    if global_ymin <= 0:
        global_ymin = 1e-10
    
    # Add some padding to limits
    global_ymin *= 0.1
    global_ymax *= 10.0
    
    # Plot UBM PSD comparison
    axes[0].loglog(norm_ubm_psd_avg.freq_r, norm_ubm_psd_avg, 'g-', linewidth=2, label='Normalized UBM')
    axes[0].loglog(unnorm_ubm_psd_avg.freq_r, unnorm_ubm_psd_avg, 'm-', linewidth=2, label='Unnormalized UBM')
    axes[0].loglog(ground_truth_ubm_psd_avg.freq_r, ground_truth_ubm_psd_avg, 'k-', linewidth=2, label='Ground Truth UBM')
    
    axes[0].set_xlabel('Wavenumber (1/m)', fontsize=15)
    axes[0].set_ylabel('Power Spectral Density', fontsize=15)
    axes[0].set_title('UBM Power Spectral Density Comparison', fontsize=16)
    axes[0].legend(loc='best', fontsize=12)
    axes[0].grid(True, which="both", ls="--", alpha=0.7)
    axes[0].set_ylim(global_ymin, global_ymax)
    
    # Plot BM PSD comparison
    axes[1].loglog(norm_bm_psd_avg.freq_r, norm_bm_psd_avg, 'g-', linewidth=2, label='Normalized BM')
    axes[1].loglog(unnorm_bm_psd_avg.freq_r, unnorm_bm_psd_avg, 'm-', linewidth=2, label='Unnormalized BM')
    axes[1].loglog(ground_truth_bm_psd_avg.freq_r, ground_truth_bm_psd_avg, 'k-', linewidth=2, label='Ground Truth BM')
    
    axes[1].set_xlabel('Wavenumber (1/m)', fontsize=15)
    axes[1].set_ylabel('Power Spectral Density', fontsize=15)
    axes[1].set_title('BM Power Spectral Density Comparison', fontsize=16)
    axes[1].legend(loc='best', fontsize=12)
    axes[1].grid(True, which="both", ls="--", alpha=0.7)
    axes[1].set_ylim(global_ymin, global_ymax)
    
    # Plot the third subplot comparing BM and UBM
    axes[2].loglog(ground_truth_bm_psd_avg.freq_r, ground_truth_bm_psd_avg, 'b-', linewidth=2, label='Ground Truth BM')
    axes[2].loglog(ground_truth_ubm_psd_avg.freq_r, ground_truth_ubm_psd_avg, 'r-', linewidth=2, label='Ground Truth UBM')
    
    axes[2].set_xlabel('Wavenumber (1/m)', fontsize=15)
    axes[2].set_ylabel('Power Spectral Density', fontsize=15)
    axes[2].set_title('BM vs UBM Power Spectral Density Comparison', fontsize=16)
    axes[2].legend(loc='best', fontsize=12)
    axes[2].grid(True, which="both", ls="--", alpha=0.7)
    axes[2].set_ylim(global_ymin, global_ymax)
    
    # Calculate RMSE statistics
    ubm_norm_rmse = calculate_rmse(results['norm_ubm'].isel(sample=random_indices), 
                                  test_data.ubm_B.isel(sample=random_indices))
    ubm_unnorm_rmse = calculate_rmse(results['unnorm_ubm'].isel(sample=random_indices), 
                                    test_data.ubm_B.isel(sample=random_indices))
    
    bm_norm_rmse = calculate_rmse(results['norm_bm'].isel(sample=random_indices), 
                                 ground_truth_bm.isel(sample=random_indices))
    bm_unnorm_rmse = calculate_rmse(results['unnorm_bm'].isel(sample=random_indices), 
                                   ground_truth_bm.isel(sample=random_indices))
    
    # Print RMSE statistics
    print(f"\nStatistics for {n_samples} randomly selected samples:")
    print("\nUBM RMSE Statistics:")
    print(f"  Normalized Model - Mean: {np.mean(ubm_norm_rmse):.4f}, Median: {np.median(ubm_norm_rmse):.4f}")
    print(f"  Unnormalized Model - Mean: {np.mean(ubm_unnorm_rmse):.4f}, Median: {np.median(ubm_unnorm_rmse):.4f}")
    print(f"  Improvement: {(np.mean(ubm_unnorm_rmse) - np.mean(ubm_norm_rmse))/np.mean(ubm_unnorm_rmse)*100:.2f}%")
    
    print("\nBM RMSE Statistics:")
    print(f"  Normalized Model - Mean: {np.mean(bm_norm_rmse):.4f}, Median: {np.median(bm_norm_rmse):.4f}")
    print(f"  Unnormalized Model - Mean: {np.mean(bm_unnorm_rmse):.4f}, Median: {np.median(bm_unnorm_rmse):.4f}")
    print(f"  Improvement: {(np.mean(bm_unnorm_rmse) - np.mean(bm_norm_rmse))/np.mean(bm_unnorm_rmse)*100:.2f}%")
    
    plt.tight_layout()
    plt.savefig('psd_comparison_three_plots.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'ubm_psd': {
            'norm': norm_ubm_psd_avg,
            'unnorm': unnorm_ubm_psd_avg,
            'ground_truth': ground_truth_ubm_psd_avg
        },
        'bm_psd': {
            'norm': norm_bm_psd_avg,
            'unnorm': unnorm_bm_psd_avg,
            'ground_truth': ground_truth_bm_psd_avg
        },
        'random_indices': random_indices
    }

# Usage example
psd_comparison = compare_psd_and_rmse(
    results, 
    test,
    n_samples=30,  # Number of random samples
    sigma=15,       # Gaussian filter strength for RMSE smoothing
    random_seed=42  # Random seed for reproducibility
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xrft
import random
import xarray as xr
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

def isotropic_spectra(data):
    """
    Calculate the isotropic power spectrum of input data.
    
    Parameters:
    -----------
    data : xarray.DataArray
        Input data array with dimensions 'i' and 'j'.
    
    Returns:
    --------
    xarray.DataArray
        The isotropic power spectrum calculated using xrft.
    """
    # Simply use the xrft function directly
    iso_psd = xrft.isotropic_power_spectrum(
        data, 
        dim=['i', 'j'], 
        detrend='constant', 
        window=True
    )
    return iso_psd

def compute_all_psds(dataarray):
    """
    Compute the PSD for each sample in a DataArray and return them all.
    
    Parameters:
    -----------
    dataarray : xarray.DataArray
        DataArray with dimensions ['sample', 'i', 'j'].
        
    Returns:
    --------
    xarray.DataArray
        All PSDs with a 'sample' dimension.
    """
    psd_list = []
    # Iterate over the sample dimension
    for s in range(len(dataarray.sample)):
        # Extract the sample as a 2D DataArray
        sample_da = dataarray.isel(sample=s)
        try:
            psd = isotropic_spectra(sample_da)
            psd_list.append(psd)
        except Exception as e:
            print(f"Error computing PSD for sample {s}: {e}")
            continue
    
    # Concatenate along the sample dimension if we have valid PSDs
    if psd_list:
        return xr.concat(psd_list, dim="sample")
    else:
        print("No valid PSDs were computed")
        return None

def plot_psd_envelope_with_mean(psd_array, label, color, ax):
    """
    Plot PSD envelope using min/max range and mean line.
    
    Parameters:
    -----------
    psd_array : xarray.DataArray
        Array containing PSDs with a 'sample' dimension.
    label : str
        Label for the plot legend.
    color : str
        Color for the plot.
    ax : matplotlib.axes.Axes
        Axes to plot on.
    """
    # Extract frequency values
    freq = psd_array.freq_r
    
    # Calculate statistics along the sample dimension
    mean_psd = psd_array.mean(dim='sample')
    min_psd = psd_array.min(dim='sample')
    max_psd = psd_array.max(dim='sample')
    
    # Plot mean line
    ax.loglog(freq, mean_psd, color=color, linewidth=2, label=f"{label} (mean)")
    
    # Plot min and max as a filled area to create envelope
    ax.fill_between(freq, min_psd, max_psd, color=color, alpha=0.3, 
                   label=f"{label} (range)")

def compare_psd_envelopes(train_data, normalized_train_data, n_samples=100, random_seed=42):
    """
    Create 2x2 subplots comparing PSD envelopes for original and normalized datasets
    using min/max envelope approach with mean line.
    """
    # Set random seed
    np.random.seed(random_seed)
    random.seed(random_seed)
    
    # Get total number of samples
    total_samples = train_data.ssh_A.shape[0]
    
    # Randomly select samples
    random_indices = np.random.choice(total_samples, size=n_samples, replace=False)
    random_indices.sort()  # Sort for consistent ordering
    
    print(f"Randomly selected {n_samples} samples for PSD analysis")
    
    # Select random samples
    ssh_A = train_data.ssh_A.isel(sample=random_indices)
    ssh_B = train_data.ssh_B.isel(sample=random_indices)
    ssh_A_norm = normalized_train_data.ssh_A_normalized.isel(sample=random_indices)
    ssh_B_norm = normalized_train_data.ssh_B_normalized.isel(sample=random_indices)
    ubm_A = train_data.ubm_A.isel(sample=random_indices)
    ubm_B = train_data.ubm_B.isel(sample=random_indices)
    ubm_A_norm = normalized_train_data.ubm_A_normalized.isel(sample=random_indices)
    ubm_B_norm = normalized_train_data.ubm_B_normalized.isel(sample=random_indices)
    
    # Create figure with 2x2 subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 14))
    
    # Compute all PSDs
    print("Computing PSDs for SSH data...")
    ssh_A_psds = compute_all_psds(ssh_A)
    ssh_B_psds = compute_all_psds(ssh_B)
    ssh_A_norm_psds = compute_all_psds(ssh_A_norm)
    ssh_B_norm_psds = compute_all_psds(ssh_B_norm)
    
    print("Computing PSDs for UBM data...")
    ubm_A_psds = compute_all_psds(ubm_A)
    ubm_B_psds = compute_all_psds(ubm_B)
    ubm_A_norm_psds = compute_all_psds(ubm_A_norm)
    ubm_B_norm_psds = compute_all_psds(ubm_B_norm)
    
    # Define datasets for each subplot
    subplot_data = [
        # (row, col, data, title)
        (0, 0, [(ssh_A_psds, "SSH_A", "red"), (ssh_B_psds, "SSH_B", "blue")], 
         'SSH Original PSD Comparison'),
        (0, 1, [(ssh_A_norm_psds, "SSH_A Normalized", "red"), (ssh_B_norm_psds, "SSH_B Normalized", "blue")], 
         'SSH Normalized PSD Comparison'),
        (1, 0, [(ubm_A_psds, "UBM_A", "red"), (ubm_B_psds, "UBM_B", "blue")], 
         'UBM Original PSD Comparison'),
        (1, 1, [(ubm_A_norm_psds, "UBM_A Normalized", "red"), (ubm_B_norm_psds, "UBM_B Normalized", "blue")], 
         'UBM Normalized PSD Comparison')
    ]
    
    # Plot each subplot
    for row, col, data_list, title in subplot_data:
        ax = axes[row, col]
        
        for psd_array, label, color in data_list:
            if psd_array is not None:
                plot_psd_envelope_with_mean(psd_array, label, color, ax)
        
        # Set plot properties
        ax.set_xlabel('Wavenumber (1/m)', fontsize=12)
        ax.set_ylabel('Power Spectral Density', fontsize=12)
        ax.set_title(title, fontsize=14)
        ax.set_xscale('log')
        ax.set_yscale('log')
        ax.legend(loc='best')
        ax.grid(True, which="both", ls="--", alpha=0.7)
    
    # Set consistent y-axis limits for all subplots
    # Find global min and max across all valid PSDs
    all_psds = [
        ssh_A_psds, ssh_B_psds, ssh_A_norm_psds, ssh_B_norm_psds,
        ubm_A_psds, ubm_B_psds, ubm_A_norm_psds, ubm_B_norm_psds
    ]
    
    global_min = float('inf')
    global_max = float('-inf')
    
    for psd in all_psds:
        if psd is not None:
            local_min = psd.min().values
            local_max = psd.max().values
            if np.isfinite(local_min) and local_min < global_min:
                global_min = local_min
            if np.isfinite(local_max) and local_max > global_max:
                global_max = local_max
    
    # Ensure global_min is positive for log scale
    if global_min <= 0:
        global_min = 1e-10
    
    # Add padding to the limits
    global_min *= 0.1
    global_max *= 10.0
    
    # Apply to all subplots
    for ax_row in axes:
        for ax in ax_row:
            ax.set_ylim(global_min, global_max)
    
    plt.tight_layout()
    plt.savefig('psd_envelope_comparison_with_mean.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'random_indices': random_indices,
        'y_range': (global_min, global_max)
    }

# Usage example
psd_envelope_results = compare_psd_envelopes(
    train_data=train,
    normalized_train_data=normalized_train,
    n_samples=100,
    random_seed=42
)
```
